#Interpolación polinomica

Supongamos que estás trabajando en un proceso químico en el cual una reacción de descomposición de
un compuesto A genera un producto B. Tienes datos experimentales que muestran la concentración de A
y la concentración de B en diferentes ejempos:

|Tiempo(min)|[A] (mol/L)|[B] (mol/L)|
|-|-|-|
|0|2.0|0|
|5|1.5|0.3|
|10|1.2|0.5|
|15|1.0|0.6|

Puedes estimar la concentración de A y B en el tiempo t = 8 minutos utilizando interpolación polinómica.


```python
import numpy as np
from scipy import interpolate
from scipy.interpolate import lagrange
import matplotlib.pyplot as plt
import sympy as sym
```

```python
x=np.array([0,5,10,15]) #variable independiente
y=np.array([2.0,1.5,1.2,1.0]) #variabel dependiente
z=np.array([0,0.3,0.5,0.6])
```

```python
p=lagrange(x,y) #<--- polinomio de lagrange
xi=8 #<----- dato a interpolar
yi=p(xi) #<----- interpolacion
print(yi)
print(p)
```

```python
xs=np.linspace(x.min(),x.max(),100) #<---- vector desde el minimo al maximo
ys=p(xs) #<---- interpolacion del vector
plt.plot(x,y,'d',xi,yi,'sm',xs,ys,'--')
plt.legend(['Datos','Interpolacion','Lagrange'])
plt.xlabel('Tiempo (min)')
plt.ylabel('Concentración A (mol/L)')
plt.title('Concentración a un tiempo')
plt.grid()
plt.show()
```

```python
q=lagrange(x,z) #<--- polinomio de lagrange
xi=8 #<----- dato a interpolar
zi=q(xi) #<----- interpolacion
print(zi)
print(q)
```

```python
xs=np.linspace(x.min(),x.max(),100) #<---- vector desde el minimo al maximo
zs=q(xs) #<---- interpolacion del vector
plt.plot(x,z,'dg',xi,zi,'sr',xs,zs,'--b')

plt.legend(['Datos','Interpolacion','Lagrange'])
plt.xlabel('Tiempo (min)')
plt.ylabel('Concentración B (mol/L)')
plt.title('Concentración a un tiempo')
plt.grid()
plt.show()
```

```python
xs=np.linspace(x.min(),x.max(),200) #<---- vector desde el minimo al maximo
ys=p(xs)

zs=q(xs) #<---- interpolacion del vector
plt.plot(x,y,'dm',xi,yi,'sr',xs,ys,'--')

plt.plot(x,z,'db',xi,zi,'sy',xs,zs,'--')
plt.legend(['Datos[A]}','Interpolacion [A]','Lagrange [A]','Datos[B]','Interpolación [B]','Lagrange[B]'])
plt.xlabel('Tiempo (min)')
plt.ylabel('Concentración (mol/L)')
plt.title('Concentración a un tiempo')
plt.grid()
plt.show()
```
