#Interpolación polinomica

Supongamos que estás trabajando en un proceso químico en el cual una reacción de descomposición de
un compuesto A genera un producto B. Tienes datos experimentales que muestran la concentración de A
y la concentración de B en diferentes ejempos:

|Tiempo(min)|[A] (mol/L)|[B] (mol/L)|
|-|-|-|
|0|2.0|0|
|5|1.5|0.3|
|10|1.2|0.5|
|15|1.0|0.6|

Puedes estimar la concentración de A y B en el tiempo t = 8 minutos utilizando interpolación polinómica.


```python
import numpy as np
from scipy import interpolate
from scipy.interpolate import lagrange
import matplotlib.pyplot as plt
import sympy as sym
```

```python
x=np.array([0,5,10,15]) #variable independiente
y=np.array([2.0,1.5,1.2,1.0]) #variabel dependiente
z=np.array([0,0.3,0.5,0.6])
```

```python
p=lagrange(x,y) #<--- polinomio de lagrange
xi=8 #<----- dato a interpolar
yi=p(xi) #<----- interpolacion
print(yi)
print(p)
```

```python
p=lagrange(x,z) #<--- polinomio de lagrange
xi=8 #<----- dato a interpolar
zi=p(xi) #<----- interpolacion
print(zi)
print(p)
```

```python
xs=np.linspace(0,15)
ys=p(xs)

plt.plot(x,y,'d',xi,yi,'sb',xs,ys,'--')
plt.legend(['Datos','Interpolacion','Lagrange'])
plt.xlabel('Tiempo')
plt.ylabel('[A](mol/L)')
plt.title('descomposición de un compuesto')
plt.grid()
plt.show()
```

```python
xs=np.linspace(0,15)
zs=p(xs)

plt.plot(x,y,'dm',xi,zi,'sy',xs,zs,'--')
plt.legend(['Datos','Interpolacion','Lagrange'])
plt.xlabel('Tiempo')
plt.ylabel('[B](mol/L)')
plt.title('descomposición de un compuesto')
plt.grid()
plt.show()
```
